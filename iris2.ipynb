```python
#43
# Step 1: Import required libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv('./Iris/Iris.csv')
print(data)
```

```python
X=data[['SepalLengthCm'  ,'SepalWidthCm'  ,'PetalLengthCm' ,'PetalWidthCm']]
print(X)
scaler=StandardScaler()
Y=data['Species']
print(Y)
X_scaled=scaler.fit_transform(X)
print(X_scaled)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, Y, test_size=0.3, random_state=42)
```

```python
# Step 4: Build Random Forest Classifier (Without PCA)
model_rf = RandomForestClassifier(n_estimators=100, random_state=42)
model_rf.fit(X_train, y_train)
```

```python
# Predicting and evaluating the model
y_pred_rf = model_rf.predict(X_test)
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print("Accuracy of Random Forest (Without PCA):", accuracy_rf)
```

```python
pca=PCA(n_components=2)
X_pca=pca.fit_transform(X_scaled)
print(X_pca)
```

```python
x_train_pca,x_test_pca,y_train_pca,y_test_pca=train_test_split(X_pca,Y,test_size=0.3,random_state=42)
```

```python
model_pca=RandomForestClassifier(n_estimators=100,random_state=42)
model_pca.fit(x_train_pca,y_train_pca)
```

```python
# Predicting and evaluating the model
y_pred_rf_pca = model_pca.predict(x_test_pca)
accuracy_rf_pca = accuracy_score(y_test_pca, y_pred_rf_pca)
print("Accuracy of Random Forest (With PCA):", accuracy_rf_pca)
```

```python
# Step 7: Compare the results
print("\nComparison:")
print("Accuracy without PCA:", accuracy_rf)
print("Accuracy with PCA:", accuracy_rf_pca)
```
