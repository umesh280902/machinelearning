```python
#Q4
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error,r2_score
```

```python
data=pd.read_csv('./advertising/Advertising Budget and Sales.csv')
print(data.columns)
```

```python
x=data[['Radio Ad Budget ($)','TV Ad Budget ($)']]
y=data['Sales ($)']
print(x,y)
```

```python
X=np.c_[np.ones(x.shape[0]),x]
print(X)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
print(x_train,x_test)
print(y_train,y_test)
```

```python
def linear_regression(X, y):
    # Transpose of X
    x_transpose = np.transpose(X)
    
    # Matrix multiplication of X^T * X
    xTx = np.dot(x_transpose, X)
    
    # Inverse of X^T * X
    xTx_inverse = np.linalg.inv(xTx)
    
    # Matrix multiplication of inverse(X^T * X) * X^T
    xtx_inverse_xTranspose = np.dot(xTx_inverse, x_transpose)
    
    # Calculate beta coefficients: beta = (X^T * X)^(-1) * X^T * y
    beta = np.dot(xtx_inverse_xTranspose, y)
    
    return beta

def linear_regression_line(beta):
    print(f"{beta[0]}+{beta[1]}*x1+{beta[2]}*x2")
    
def predictions(X,beta):
    return np.dot(X,beta)
```

```python
# Calculate beta coefficients
beta = linear_regression(x_train, y_train)

# Print the regression equation
linear_regression_line(beta)

# Make predictions on X_test using the beta coefficients
y_pred = predictions(x_test, beta)

# Print the predicted values
print("Predicted Sales:", y_pred)
```

```python
rmse=root_mean_squared_error(y_test,y_pred)
print(rmse)

r2=r2_score(y_test,y_pred)
print(r2)

print(y_test.values)
print(y_pred)
```
