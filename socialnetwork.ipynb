```python
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score,f1_score
```

```python
data=pd.read_csv('./Social network dataset/Social_Network_Ads.csv')
print(data.info())
```

```python
le=LabelEncoder()
data['Gender']=le.fit_transform(data['Gender'])
```

```python
features=data[['Gender','Age','EstimatedSalary']].values
target=data['Purchased']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(features,target,test_size=0.3,random_state=42)
```

```python
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
model=SVC(kernel='linear',random_state=42)
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
#19,20,21
cm = confusion_matrix(Y_test, y_pred)
print("Confusion Matrix:")
print(cm)

# Calculate accuracy
accuracy = accuracy_score(Y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")

# Calculate recall
recall = recall_score(Y_test, y_pred)
print(f"Recall: {recall:.2f}")

# Calculate precision
precision = precision_score(Y_test, y_pred)
print(f"Precision: {precision:.2f}")
```

```python
#22,26
# Step 2: Use SVM to classify using RBF Kernel
rbf_model = SVC(kernel='rbf', random_state=42)
rbf_model.fit(X_train, Y_train)

# Step 3: Create a confusion matrix and evaluate using F1-Measure
y_pred = rbf_model.predict(X_test)
cm = confusion_matrix(Y_test, y_pred)
print(cm)
f1 = f1_score(Y_test, y_pred)
print(f"F1 score:{f1:.2f}")
```
